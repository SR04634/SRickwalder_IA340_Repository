```python
import matplotlib.pyplot as plt
crime_by_type = df.groupby('Crm Cd Desc').size().reset_index(name='Count')
crime_by_type.head()
crime_by_area = df.groupby('AREA NAME').size().reset_index(name='Count')
crime_by_area.head()
crime_by_area_type = df.groupby(['AREA NAME', 'Crm Cd Desc']).size().reset_index(name='Count')
crime_by_area_type.head()

top_areas = crime_by_area.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_areas['AREA NAME'], top_areas['Count'], color='skyblue')
plt.xlabel('Number of Crimes')
plt.ylabel('Area')
plt.title('Top 10 Areas by Number of Crimes')
plt.gca().invert_yaxis()  # highest at top
plt.show()
```

This plot give you the top 10 areas by number of crimes. The data shows what cities have the lowest number of crimes (Wilshire) and the highest number of crimes (Central).
